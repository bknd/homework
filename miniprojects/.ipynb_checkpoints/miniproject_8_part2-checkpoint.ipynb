```python
import pandas as pd
```

```python
pokemon = pd.read_csv('https://stepik.org/media/attachments/lesson/363874/Pokemon.csv')
```

```python
pokemon.shape
```

```python
pokemon.head()
```

```python
#изменим названия исходных столбцов для удобства
pokemon.columns = pokemon.columns.str.replace('\. ', '_').str.replace(' ', '_').str.lower()
```

```python
#переименуем колонку
pokemon.rename(columns={'#': 'id'}, inplace=True)
```

```python
pokemon.head()
```

```python
#выясним, сколько в каком поколении легендарных покемонов
legends = pokemon.groupby('generation') \
                 .agg({'legendary': 'value_counts'})
```

```python
legends.head()
```

```python
#переименуем колонку, чтобы не было путаницы, и поместим уровень индекса legendary в уровень столбцов
legends_unstacked = legends.rename(columns={'legendary': 'legendary_count'}) \
                           .unstack(level='legendary')
```

```python
legends_unstacked.head()
```

```python
#узнаем, среди каких типов покемонов и в каком поколении больше всего легендарных
pokemon.groupby(['generation', 'type_1']) \
       .agg({'legendary': 'sum'}) \
       .unstack(level='type_1')
```

```python
#получим новые данные о способностях супергероев
superheroes = pd.read_csv('https://stepik.org/media/attachments/lesson/363874/superheroes_power_matrix.csv')
```

```python
superheroes.shape
```

```python
superheroes.head()
```

```python
#преобразуем датафрейм в длинный формат
superheroes_long = superheroes.melt(id_vars='Name', var_name='superpower')                              
```

```python
#отфильтруем значения и объединим суперспособности для каждого героя
superheroes_powers = superheroes_long.query('value == True') \
                                     .groupby('Name') \
                                     .superpower \
                                     .apply(list) \
                                     .to_frame() \
                                     .reset_index()
```
